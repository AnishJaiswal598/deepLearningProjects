```python
import pandas as pd
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
# Since the dataset has a 28*28 grid we will have that many no of inputs toward the neural networks which can be too much and 
# hence we need to create a hidden layer that can reduce our no of input...
# We will first reshape our data to get it in one single line
X_train_flattened=X_train.reshape(len(X_train),-1)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened=X_test.reshape(len(X_test),-1)
X_test_flattened.shape
```

```python
X_train_flattened
```

```python
# Creating a simple neural network with 784 input layers and 10 output layers

# here Dense means all the input layer are connected to all the output layer
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

# Optimizer does here the job of finding the global maxima and the method i.e the optimizer used is adam
# Since we have categorical value we use it as a loss parameter, even we can use mean squared error also and anything else
# Look at tensorflow
# metrics=accuracy is because we want to make it more accurate which is our end goal
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
# accuracy increases with the no of epochs increases and loss decreases 
```

```python
# lets scale down our matrix to lower no and do the process again
X_train_flattened=X_train_flattened/255
X_test_flattened=X_test_flattened/255
```

```python
# Creating a simple neural network with 784 input layers and 10 output layers

# here Dense means all the input layer are connected to all the output layer
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

# Optimizer does here the job of finding the global maxima and the method i.e the optimizer used is adam
# Since we have categorical value we use it as a loss parameter, even we can use mean squared error also and anything else
# Look at tensorflow
# metrics=accuracy is because we want to make it more accurate which is our end goal
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
# We therefore see that accuracy of our model increased
```

```python
# Now lets do testing on our test_dataset
model.evaluate(X_test_flattened,y_test)
```

```python
# we got a accuracy of 92.38%
plt.matshow(X_test[0])
y_predicted=model.predict(X_test_flattened)
```

```python
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
# So we see that we have predicted it correctly
plt.matshow(X_test[7])
np.argmax(y_predicted[7])
```

```python
y_predicted_max=[np.argmax(i) for i in y_predicted]
conf_matrix=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_max)
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# Adding hidden layer of neural network

# here Dense means all the input layer are connected to all the output layer
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

# Optimizer does here the job of finding the global maxima and the method i.e the optimizer used is adam
# Since we have categorical value we use it as a loss parameter, even we can use mean squared error also and anything else
# Look at tensorflow
# metrics=accuracy is because we want to make it more accurate which is our end goal
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
# We see that accuracy went up to 98.4%
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted2=model.predict(X_test_flattened)
y_predicted_max_withHid=[np.argmax(i) for i in y_predicted2]
conf_matrix2=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_max_withHid)
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix2,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# We can further just provide the raw matrix instead of creating a flattened list
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

# Optimizer does here the job of finding the global maxima and the method i.e the optimizer used is adam
# Since we have categorical value we use it as a loss parameter, even we can use mean squared error also and anything else
# Look at tensorflow
# metrics=accuracy is because we want to make it more accurate which is our end goal
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predicted3=model.predict(X_test)
y_predicted_max_withHid=[np.argmax(i) for i in y_predicted3]
conf_matrix3=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_max_withHid)
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix3,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
